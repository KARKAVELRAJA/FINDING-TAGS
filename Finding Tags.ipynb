# Finding Tags

```python
# Importing Necessary Libraries

from bs4 import BeautifulSoup
import requests
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# Web Scraping

url_link     = "https://en.wikipedia.org/wiki/Virat_Kohli"
url_response = requests.get(url_link)
url_response.text
```

```python
# Passing the HTML_Code into Beautiful_soup

soup = BeautifulSoup(url_response.text,"html.parser")
soup
```

```python
# Text Pre-processing

text = soup.get_text(strip=True)
text
```

```python
# Tokenization

tokens = []

for i in text.split():
    tokens.append(i)
    
tokens
```

```python
# Removing Stop_Words

print("Stop_Words :",stopwords.words("english"))
cleaned_tokens = tokens[:] 
for i in tokens:
    if i in stopwords.words("english"):
        cleaned_tokens.remove(i)
print("-"*125)
print("stop_words_removed_tokens:",cleaned_tokens) # stop_words_removed_tokens
```

```python
# Finding Tags

freq = nltk.FreqDist(cleaned_tokens)
freq
```

```python
for key,value in freq.items():
    print(str(key)+ ":" +str(value))
```

```python
freq.plot(8,cumulative=False)
```
